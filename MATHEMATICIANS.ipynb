```
#Introduction
print('WELCOME TO THE WORLD OF SCIENCES!\nLET US CHECK IF YOU COULD STILL REMEMBER THE PEOPLE BEHIND SCIENCE!')
print('\nLet us see how much you remember about the greatest Scientist in the history!\nA little bit of your school memories will be tested.')
print('\nBut before that let me know a little information about yourself, so I would know how do I call you.\n')
name = input( 'What is your name?' )
print("Hello, " + name.title() + "! " + "Are you ready to take the exam? If so, then, let's begin!")

#Questions:
def QUIZ():
  questions = ['\n1. I am the Father of Mathematics.\n',
              '\n2. I am the Father of Geometry.\n',
              '\n3. I am the Father of Biology.\n',
              '\n4. I am the first recorded Mathematician who used pi as Mathematical symbol.\n',
              '\n5. The Cartesian Plane is named after me, who am I?\n',
              '\n6. One of my greatest constributions in Mathematics is the famous Pythagorean Theorem.\n',
              '\n7.He is known as the Father of Modern Physics.\n',
              '\n8. I am well-known for I formulated the Law of Gravitation and the Three Laws of Motion.\n',
              '\n9. I established zero as a number and defined its mathematical properties; discovered the formula for solving quadratic equations.\n',
              '\n10. I am the author of one of the most famous books in history, On the Origin of Species.\n']

  SCIENTISTS = ['Archimedes','Euclid','Aristotle','William Jones','Rene Descartes','Pythagoras','Albert Einstein','Isaac Newton','Brahmagupta','Charles Darwin']

  c = 0
  score = 0
  for question in questions:
    print(question)

    ans = input('Please enter your answer:')

    if ans.title() == SCIENTISTS[c]:
      print("\nVery Good!\n")
      score = score + 1
    else:
      print("\nSorry you're wrong!\n")

    c +=1

#scoring
  print("\nYour Final Score is:{0} out of 10".format(score))

  if score == 10:
    print("\nGreat Job " + name.title() + "! CONGRATULATIONS!")
  elif score == 9:
    print("\nVery Good" + name.title() + "! CONGRATULATIONS!")
  elif score == 8:
    print("\nThat's good" + name.title() + "! CONGRATULATIONS!")
  else:
    print("\nSorry " + name.title() + ", but I think you need to refresh your mind!")


QUIZ()







  

```
